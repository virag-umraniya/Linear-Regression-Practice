## 2. Domain Analysis - Ecommerce Customers Dataset

This dataset contains information about **Ecommerce platform users**, capturing their behavior on the platform and key metrics tied to spending patterns.

### i. Dataset Description
The dataset includes the following features:

- **Email**: Customer email address (identifier).
- **Address**: Full physical address of the customer.
- **Avatar**: Categorical variable representing user's profile avatar color.
- **Avg. Session Length**: Average session length on the website (in minutes).
- **Time on App**: Average time spent on the mobile app (in minutes).
- **Time on Website**: Average time spent on the website (in minutes).
- **Length of Membership**: Number of years the customer has been a member.
- **Yearly Amount Spent**: Target variable â€” total amount of money (in USD) the customer spends in a year.

### ii. Objective
The primary goal is to analyze **user behavior** to determine:
- What drives higher spending?
- How usage patterns differ between app vs. website.
- Which features correlate strongly with the **Yearly Amount Spent**.

### iii. Data Types
- **Categorical**: Avatar
- **Textual**: Email, Address
- **Numerical**: All other columns

### iv. Use Cases
- Customer segmentation based on spending patterns.
- Personalized marketing strategies.
- Feature importance for predictive modeling.

## 7. Feature Extraction

```python
from sklearn.decomposition import PCA

pca = PCA()
pc = pca.fit_transform(df)

# Cumulative explained variance
cum_var = np.cumsum(pca.explained_variance_ratio_)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cum_var)+1), cum_var * 100, marker='o', linestyle='-')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('Explained Variance by PCA')
plt.grid(True)
plt.show()
```

## 8. Split the data

```python
X = df.iloc[:,:4]
X
```

```python
y = df[['Yearly Amount Spent']]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
print("X_train size:",X_train.shape)
print("X_test size:",X_test.shape)
print("y_train size:",y_train.shape)
print("y_test size:",y_test.shape)
```

## 9. Model Creation

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train, y_train)
```

```python
y_predict = LR.predict(X_test)
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
y_test.head()
```

```python
y_predict
```

## 10. Model Evaluation

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
y_predict = LR.predict(X_test)
print("R2 score:",r2_score(y_test, y_predict))
print("MAE:",mean_absolute_error(y_test, y_predict))
print("MSE:",mean_squared_error(y_test, y_predict))
```

## 11. Suggestion to the Problems

```python
from sklearn import metrics
coeff_df = pd.DataFrame(LR.coef_.T, index=X.columns, columns=['Coefficient'])
print(coeff_df)
```

### 1. What drives higher spending?

According to the linear regression model, the following features drive higher spending:

- **Length of Membership (61.30)** is the strongest driver. Customers who have been members for longer tend to spend more.
- **Time on App (38.81)** is also a significant driver. Users who spend more time on the app tend to have higher spending.
- **Avg. Session Length (25.83)** shows a moderate positive impact.
- **Time on Website (0.28)** has very little impact on spending.

**Conclusion**: Customers who have been members longer and actively use the app with longer session durations are more likely to spend more.

---

### 2. How usage patterns differ between app vs. website?

The coefficient for **Time on App** is significantly higher than **Time on Website**:

- Time on App: 38.81
- Time on Website: 0.28

**Conclusion**: Spending time on the app is much more strongly associated with higher yearly spending compared to the website. This suggests the app is more effective at converting user activity into purchases.

---

### 3. Which features correlate strongly with the Yearly Amount Spent?

Based on the model coefficients:

1. Length of Membership â†’ 61.30

3. Avg. Session Length â†’ 25.83
4. Time on Website â†’ 0.28

**Conclusion**: The strongest correlation is with Length of Membership, followed by Time on App. Time on Website shows almost no correlation.
